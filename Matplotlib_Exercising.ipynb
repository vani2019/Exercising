```
x=[1,2,3,4]
y=[19,25,39,41]
plt.plot(x,y);
```

```
## method 1

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y);
```

```
## method 2

fig = plt.figure(figsize=(4,3))
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y);
```

```
## method 3  

fig, ax = plt.subplots(figsize=(5,4))  ## (width, height)
ax.plot(x,y)
ax.set(title="Simple Plot", xlabel="X-axis", ylabel="Y-axis")
fig.savefig("my_matplot.png")
```

## Making figures with numpy arrays

```
x = np.linspace(0, 10, 100)
x[:10]
```

```
##  default line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```
##  scatter plot

fig, ax = plt.subplots()
ax.scatter(x, x**2);
```

```
## scatter plot

fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```
## Bar plot
## make a plot from dict

butterprices = {"Almond butter": 8, "Peanut butter": 4, "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(butterprices.keys(), butterprices.values())
ax.set(title="Butter Prices", ylabel="Price");
```

```
## bar plot (horizontal bar)

fig, ax = plt.subplots()
ax.barh(list(butterprices.keys()), list(butterprices.values()));
```

```
## Histogram plot

x=np.random.randn(1000)  ## normal distribution sample
fig,ax = plt.subplots()
ax.hist(x);
```

## Subplots exploration

```
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10), color="orange")
ax3.bar(butterprices.keys(), butterprices.values(), color="green")
ax4.hist(x, color="red");
```

```
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# plot to each different axis
ax[0, 0].plot(x, x/2, color="orange")
ax[0, 1].scatter(np.random.random(10), np.random.random(10), color="red")
ax[1, 0].bar(butterprices.keys(), butterprices.values(), color="cyan")
ax[1, 1].hist(x, color="blue");
```

## Plotting from Pandas Dataframes

```
carsales= pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales.csv")
carsales
```

```
carsales['Price']=carsales['Price'].str.replace('[\$\,\.]', '')
```

```
carsales['Price']=carsales['Price'].str[:-2]
```

```
carsales
```

```
carsales['sale date']=pd.date_range('1/1/2020', periods=len(carsales))
carsales
```

```
carsales['Price']=carsales['Price'].astype(int)
carsales['Total Sales']=carsales['Price'].cumsum()
carsales
```

```
carsales.plot(x='sale date', y='Total Sales');
```

```
carsales.plot(x='Odometer (KM)', y='Price', kind='scatter');
```

```
heartdis = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv")
heartdis
```

```
# distribution of age - histogram
heartdis.age.plot(kind='hist');
```

```
# distribution of all columns - histogram

heartdis.plot(kind='hist', subplots=True, figsize=(10,30));
```

```
# analysis on patients of age > 50

over50 = heartdis[heartdis['age']>50]
over50
```

```
# plot of age, chol with target (to classify) for age>50

fig, ax = plt.subplots(figsize=(10, 6))
over50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
```

```
fig, ax = plt.subplots(figsize=(10, 6))
sc=ax.scatter(x=over50['age'], y=over50['chol'], c=over50['target'])
ax.set(title="Heart Disease and Cholestorol levels for Aged people", xlabel="Age", ylabel="Cholestorol")
ax.legend(*sc.legend_elements(), title='Target');
# add a horizontal line at mean
ax.axhline(over50['chol'].mean(), linestyle='--');
```

```
## subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)
sc=ax0.scatter(x=over50['age'], y=over50['chol'], c=over50['target'])
ax0.set(title='Heart Disease and Cholestorol',  ylabel='Cholestorol')
ax0.legend(*sc.legend_elements(), title='Target')
ax0.axhline(over50['chol'].mean(), linestyle='--');

sc1=ax1.scatter(x=over50['age'], y=over50['thalach'], c=over50['target'])
ax1.set(title='Heart Disease and Max Heart Rate', xlabel='Age', ylabel='Max Heart Rate')
ax1.legend(*sc1.legend_elements(), title='Target')
ax1.axhline(over50['thalach'].mean(), linestyle='--');

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
